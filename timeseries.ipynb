```python
# Decompose a time series dataset to identify trends and seasonal pattersn

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# time series data (DD-MM-YYYY and sale data)
date_range = pd.date_range(start='2020-01-01', periods = 26, freq='ME')
sales_data = [
    120, 130, 150, 160, 180, 200, 220, 210, 200, 190, 180, 160, 180,
    150, 140, 160, 170, 190, 210, 230, 220, 210, 200, 190, 180, 190
]

df = pd.DataFrame({
    "Date" : date_range,
    "Sales": sales_data
})

df.set_index("Date", inplace=True)

# decompose the time series here!
decomposition = seasonal_decompose(df["Sales"], model= "multiplicative")

# plot the decomposition here! ( to visualize the our output here !)
decomposition.plot()
plt.show()
```
