```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# to create parameters for binomial distribution
n = 100 # number of trails
p = 0.5 # probabilty of success 50%

# simulate binomial ditribution 
binom_dist = np.random.binomial(n, p, 1000)

# plot binomial distribution 
plt.hist(binom_dist, bins=30, alpha=0.7, label="Binomial")

#plot normal distribution for comparison both data

X = n * p
sigma = np.sqrt(n * p * (1 - p))
normal_dist = np.random.normal(X, sigma, 1000)
plt.hist(normal_dist, bins=30, alpha=0.7, label="Normal")

# applying the plot methods in the distibution ( histograms )
plt.legend()
plt.title("Binomial vs Normal")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

```
