```python
# Linear Regression is one of the simplest and mostly used algorithm in ML and Statistics
# so it is relationship b/w a dependent variable (output) and one independent variable (input)

# Y = b0 + b1X ( the sum of square differences b/w the predicted and actual values know as least square method)

# Y = it is dependent variable (predict value)
# X = independent variable ( input value)
# b0 = intercept 
# b1 = slope ( rate at which Y changes with X)


```

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression

```

```python
# Data ( Square Footage and Price for House)
X = np.array([1500, 2000, 25000, 3000, 3500]).reshape(-1, 1) #  independent varaible - input data
Y = np.array([30000, 40000, 50000, 60000, 70000]) # dependedent variable 

# create and fit the Model 
model = LinearRegression() # Pre-defined Model
model.fit(X, Y) # fit method on Linear Regression

# Predict the Model here
Y_Pred = model.predict(X)

# visualizing the linear regressionr result or data

plt.scatter(X, Y, color="blue")
plt.plot(X, Y_Pred, color="red")
plt.title("House price Prediction")
plt.xlabel("Square Footage")
plt.ylabel("House Price")
plt.show()

# co-efficient of the Linear Regression Equation
print("Intercept (b0):", model.intercept_)
print("Slope (b1):", model.coef_[0])


```
