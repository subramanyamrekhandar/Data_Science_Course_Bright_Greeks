```python
#build a simple linear regresion model to predict housing price based on square footage
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# sample data
square_footage = np.array([1500, 1700, 1800, 2000, 2200, 2500]).reshape(-1, 1)
prices = np.array([300000, 340000, 360000,400000, 420000, 500000])

# create and train the model
model = LinearRegression()
model.fit(square_footage, prices )

#predict prices
pridicted_prices = model.predict(square_footage)

# plot the data and regression line in the visualize data here
plt.scatter(square_footage, prices, color = 'blue')
plt.plot(square_footage, pridicted_prices, color = 'red')
plt.title("Linear Regression: Housing Price Vs Square Footage")
plt.xlabel("Square Footage")
plt.ylabel("price")
plt.show()



```
